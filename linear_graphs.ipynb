```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('average-height-by-year-of-birth.csv')
data = data.drop(labels='Code', axis=1)
us_data = data[data.Entity == 'United States']

year = [x for x in us_data.iloc[:,1]]
male_data = [x for x in us_data.iloc[:,2]]
female_data = [y for y in us_data.iloc[:,3]]

plt.plot(year, male_data, label = 'Male')
plt.plot(year, female_data, label = 'Female')
plt.title('Average Height 1896-1996')
plt.xlabel('Year')
plt.ylabel('Average Height (cm)')
plt.legend()
plt.show




```
